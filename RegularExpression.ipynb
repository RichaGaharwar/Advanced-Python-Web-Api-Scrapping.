````python
# write a function to mask email address
# sample: abcd@gmail.com --> a####d@gmail.com

def mask_email (email):
  if "@" in email:
    name, domain = email.split("@")
    return f"{name[0]}####{name[-1]}@{domain}"
  else:
    return "Invalid email"
````

````python
email1 = "rsg@gmail.com"
````

````python
mask_email(email1)
````

````python
mask_email("richa.com")
````

Regex - Regular Expression come into action!

````python
import re
def is_valid_email(input):
  email_pattern = "^\w+([\.-]?\w+)*@\w+([\.-]?\w+)*(\.\w{2,3})+$"
  result = re.search(email_pattern,input)
  print(result)
  # scan through string looking for the first location where the pattern is found
  if result:
    return True
  else:
    return False
````

````python
is_valid_email(email1)
````

## Quantifiers

available quantifiers:
```
* - 0 or more

+ - 1 0r more

? - 0 or one

{3} - exact number

{3,4} - range of numbers (min, max)

````python
regex = "^\w+([\.-]?\w+)*@\w+([\.-]?\w+)*(\.\w{2,3})+$"
````

1.   ^ means start & $ means end

2.   \w+ means alphanumeric, one or more times

3.   ([\.-]?W+)* - zero or more charecters after
@

4.   @\w= - 1 or more word charecters after @

5.   ([\.-]?\w+)* - zero or more times - either dot or dash, followed by words.

6.   (.\w{2,3})+ - matches followed by 2 or 3 charecters ie. .com or .in etc





````python
regex = "^\w+([\.-]?\w+)*@\w+([\.-]?\w+)*(\.\w{2,3})+$"

reg_pattern = re.compile(r"""
                          ^\w+([\.-]?\w+)*   # starts with \w+ - user
                          @                  # single @ sign
                          \w+([\.-]?\w+)*    # domain name
                          (\.\w{2,3})+$      # .com, .ac, .in, .co
                          """, re.VERBOSE | re.IGNORECASE)
result = reg_pattern.match("abcd@iisc.ac.in")
print(result.string)
print(result)
````

## Important functions

1.   match - checks for a match only at the beginning the string

2.   search - locates the pattern in the string

3.   findall - finds all the occurrences in the string

4.   finditer - return an interior yielding matche objects over all non overlapping matches.







````python
pattern = ("\d{3}-\d{3}-\d{4}")
print(re.match(pattern, data))
````

works only if the pattern is in the begining

````python
print(re.search(pattern, data))
````

gives the first occurance

````python
print(re.findall(pattern, data))
````

list of all phone numbers from text

````python
numbers = re.finditer(pattern, data)
for i, num in enumerate(numbers):
  print(num)
  if (i==5):
    break
````

````python
numbers = re.finditer(pattern, data)
for i, num in enumerate(numbers):
  print(num.group(), num.start(), num.end())
  if (i==5):
    break
````

gives numbers with start and end index
